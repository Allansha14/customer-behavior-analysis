## 5. Handle Missing Values (Review Rating)

Purpose: Fill missing review ratings category-wise

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'] \
    .transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

## 6. Standardize Column Names

Purpose: Make columns SQL-friendly

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
df.columns
```

## 7. Rename Column

Purpose: Simplify column name

```python
df.rename(columns={'purchase_amount_usd': 'purchase_amount'}, inplace=True)
df.columns
```

## 8. Feature Engineering – Age Group

Purpose: Customer segmentation

```python
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)

df[['age','age_group']].head(10)
```

## 9. Feature Engineering – Purchase Frequency Days

Purpose: Convert text frequency → numeric

```python
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

## 10. Validate Discount Column & Drop Unused Column

Purpose: Remove unnecessary data

## 12. Connect Jupyter → MySQL

Purpose: Send cleaned data into SQL database

```python
from sqlalchemy import create_engine

engine = create_engine(
    "mysql+pymysql://root:1234@localhost:3306/customer_behavior"
)
```

## 13. Load Data into MySQL Table

Purpose: Store processed data for SQL usage

```python
df.to_sql(
    name='customer',
    con=engine,
    if_exists='replace',
    index=False
)

print("Data successfully loaded into MySQL database")
```

```python
import matplotlib.pyplot as plt
from sqlalchemy import create_engine
```

```python
query = """
SELECT
    category,
    SUM(`purchase_amount_(usd)`) AS total_sales
FROM customer
GROUP BY category
ORDER BY total_sales DESC;
"""

df_sales = pd.read_sql(query, engine)
df_sales
```

```python
plt.figure(figsize=(8, 5))
plt.bar(df_sales['category'], df_sales['total_sales'])
plt.xlabel("Product Category")
plt.ylabel("Total Sales (USD)")
plt.title("Total Sales by Category")
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 6))
plt.pie(
    df_sales['total_sales'],
    labels=df_sales['category'],
    autopct='%1.1f%%',
    startangle=140
)
plt.title("Sales Contribution by Category")
plt.show()
```

```python
# Fetch average review rating by category
query_rating = """
SELECT
    category,
    AVG(review_rating) AS avg_review_rating
FROM customer
GROUP BY category
ORDER BY avg_review_rating DESC;
"""

df_rating = pd.read_sql(query_rating, engine)
df_rating
```

```python
plt.figure(figsize=(8, 5))
plt.bar(df_rating['category'], df_rating['avg_review_rating'])
plt.xlabel("Product Category")
plt.ylabel("Average Review Rating")
plt.title("Average Customer Review Rating by Category")
plt.xticks(rotation=30)
plt.ylim(0, 5)
plt.tight_layout()
plt.show()
```
